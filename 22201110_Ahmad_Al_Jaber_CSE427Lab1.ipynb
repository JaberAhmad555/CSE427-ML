```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#1
a = np.random.randint(1, 101, 20)
b = np.random.randint(1, 101, 20)

print("a:",a)
print("b:",b)
```

```python
#2
a_matrix = a.reshape(5, 4)
b_matrix = b.reshape(5, 4)

print("a:",a_matrix)
print("b:",b_matrix)
```

```python
#3
scalar_multiplication_result = a_matrix * b_matrix
print("scalar result:", scalar_multiplication_result)
```

```python
#4
b_transposed = b_matrix.T
c = np.matmul(a_matrix, b_transposed)
print("C:", c)
```

```python
#5
c_sliced_columns = c[:, 2:4]
print("Columns 2 and 3 of matrix c:\n", c_sliced_columns)
c_whole_sliced = c[:, :]
print("Whole matrix c using slicing:\n", c_whole_sliced)
```

```python
#6
largest_element = np.max(c)
largest_element_index_flat = np.argmax(c)
largest_element_index= np.unravel_index(largest_element_index_flat, c.shape)
print(f"Largest element in c: {largest_element}")
print(f"Index of the largest element: {largest_element_index}")
```

```python
#7
c_flattened = c.flatten()
print("Matrix c converted to a 1D array:\n", c_flattened)
print("Original matrix c:\n", c)
```

```python
#1
df = pd.read_csv('/content/subject_scores.csv')
```

```python
#2
print(df.head())
```

```python
#3
print(df.info()) #detail
print(df.describe()) #statistical
```

```python
#4
print(df.isnull().sum())
```

```python
#5
for col in df.columns:
    if pd.api.types.is_numeric_dtype(df[col]):
        if df[col].isnull().sum() > 0:
            mean_value = df[col].mean()
            df[col].fillna(mean_value, inplace=True)

print("\nNull values after replacing with mean:")
print(df.isnull().sum())
```

```python
#6
math_scores = df['Math']
print("\nMath scores of all students:")
print(math_scores)
```

```python
temperature_dhaka = [25, 34, 21, 45, 28, 6, 43, 18, 7, 2]
humidity_dhaka = [28, 25, 29, 20, 26, 50, 19, 29, 52, 55]
study_hours = [2, 3, 4, 4, 5, 6, 7, 7, 8, 9, 9, 10, 11, 11, 12]
marks = [6, 10, 15, 20, 34, 44, 55, 60, 55, 67, 70, 80, 90, 99, 100]
subjects = ['Maths', 'English', 'Science', 'Physics', 'Computer']
subject_marks = [89, 90, 45, 78, 99]
```

```python
#1
plt.subplot(2, 2, 1)
plt.scatter(temperature_dhaka, humidity_dhaka, marker='*', color='red')
plt.title('Temperature vs. Humidity in Dhaka')
plt.xlabel('Temperature (Celsius)')
plt.ylabel('Humidity (%)')
plt.grid(True)
```

```python
#2
plt.subplot(2, 2, 2)
plt.plot(study_hours, marks, marker='o', linestyle='-')
plt.title('Study Hours vs. Marks')
plt.xlabel('Study Hours')
plt.ylabel('Marks')
plt.grid(True)
```

```python
#3
plt.subplot(2, 2, 3)
plt.barh(subjects, subject_marks)
plt.title('Scores in Different Subjects (Horizontal)')
plt.xlabel('Marks')
plt.ylabel('Subjects')
```

```python
#4
plt.subplot(2, 2, 4)
colors = ['blue', 'green', 'red', 'purple', 'orange']
plt.bar(subjects, subject_marks, color=colors)
plt.title('Scores in Different Subjects (Vertical)')
plt.xlabel('Subjects')
plt.ylabel('Marks')
```

```python
#5
plt.subplot(2, 2, 1)
plt.scatter(temperature_dhaka, humidity_dhaka, marker='*', color='red')
plt.title("Temperature vs Humidity")

plt.subplot(2, 2, 2)
plt.plot(study_hours, marks)
plt.title("Study Hours vs Marks")

plt.subplot(2, 2, 3)
plt.barh(subjects, subject_marks)
plt.title("Scores in Each Subject")

plt.subplot(2, 2, 4)
plt.bar(subjects, subject_marks, color=['blue', 'green', 'red', 'purple', 'orange'])
plt.title("Scores in Each Subject (Vertical)")

plt.tight_layout()
plt.show()
```

```python
import pandas as pd
from sklearn.datasets import load_digits
digits = load_digits()
```

```python
dir=(digits)
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
```

```python
plt.gray()
for i in range (4):
  plt.matshow(digits.images[i])
```

```python
df = pd.DataFrame(digits.data)
df.head()
```

```python
df['target']=digits.target
df.head()
```

```python
from sklearn.model_selection import train_test_split
X = df.drop('target',axis='columns')
y = df.target
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.2)
```

```python
len(X_train)
```

```python
from sklearn.ensemble import RandomForestClassifier
model = RandomForestClassifier(n_estimators=40)#estimator dile tree count baraate parbo and then accuracy can also increase
model.fit(X_train, y_train)
```

```python
model.score(X_test,y_test) # here you go accuracy increased
```

```python
#y_test=truth and y_predicted is predicted
y_predicted=model.predict(X_test)
```

```python
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test, y_predicted)
import seaborn as sn
plt.figure(figsize=(10,7))
sn.heatmap(cm, annot=True)
plt.xlabel('Predicted')
plt.ylabel('Truth')
```
